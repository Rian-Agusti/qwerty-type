```python
combine = pd.concat([ddos_icmp,ddos_udp,dos_icmp],ignore_index=True)
```

```python
combine.columns.values
```

```python
n = combine.iloc[:,7: 76]
```

```python
n
```

```python
m = combine.iloc[:,83: 84]
```

```python
m
```

```python
from sklearn.model_selection import train_test_split
```

```python
n_train,n_test,m_train,m_test = train_test_split(n,m, test_size = 0.2, random_state = 42)
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
monka = DecisionTreeClassifier(criterion='entropy', splitter = 'random')
monka.fit(n_train,m_train)
```

```python
m_pred = monka.predict(n_test)
```

```python
m_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(m_test,m_pred)
```

```python
accuracy
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize = (10, 7))
tree.plot_tree(monka, feature_names = n.columns.values, class_names = np.array([ 'DoS ICMP Flood','DDos ICMP Flood','DDoS UDP Flood']), filled = True)
plt.show()
```

```python
import seaborn as mnm
from sklearn import metrics
label = np.array([ 'ddos_icmp','ddos_udp','dos_icmp'])
```

```python
import matplotlib.pyplot as plt
conf_matrix = metrics.confusion_matrix(m_test, m_pred)
plt.figure(figsize=(14, 14))
mnm.heatmap(conf_matrix, annot=True, xticklabels=label, yticklabels=label)
plt.xlabel('Prediksi')
plt.ylabel('Fakta')
plt.show()
```
